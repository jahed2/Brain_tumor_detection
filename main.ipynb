```python
import numpy as np
import os
import cv2
from PIL import Image
from sklearn.model_selection import train_test_split

image_directory = 'datasets/'

no_tumor_images = os.listdir(image_directory + 'no/')
yes_tumor_images = os.listdir(image_directory + 'yes/')

dataset = []
label = []

for image_filename in no_tumor_images:
    image_path = os.path.join(image_directory, 'no', image_filename)
    image = cv2.imread(image_path)  # Read the image using OpenCV
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    image = Image.fromarray(image)  # Create a PIL Image object
    image = image.resize((64, 64))
    dataset.append(np.array(image))
    label.append(0)  # Label 0 for 'no' tumor images

for image_filename in yes_tumor_images:
    image_path = os.path.join(image_directory, 'yes', image_filename)
    image = cv2.imread(image_path)  # Read the image using OpenCV
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    image = Image.fromarray(image)  # Create a PIL Image object
    image = image.resize((64, 64))
    dataset.append(np.array(image))
    label.append(1)  # Label 1 for 'yes' tumor images

dataset = np.array(dataset)
label = np.array(label)

# Perform stratified train-test split
X_train, X_test, y_train, y_test = train_test_split(
    dataset, label, test_size=0.2, random_state=42, stratify=label
)

X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Define the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

model evaluate 

```python
loss, accuracy = model.evaluate(X_test,y_test)
```

```python
print("Test loss:", loss)
print("Test accuracy", accuracy)
```

```python
model.save('my_model.h5')
```

```python
# Make predictions on new data
new_data_directory = 'pred/'
new_data_filenames = os.listdir(new_data_directory)

new_data = []
for image_filename in new_data_filenames:
    image_path = os.path.join(new_data_directory, image_filename)
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = Image.fromarray(image)
    image = image.resize((64, 64))
    new_data.append(np.array(image))

new_data = np.array(new_data)
new_data = new_data.astype('float32') / 255.0

# Load the saved model
loaded_model = tf.keras.models.load_model('my_model.h5')

# Make predictions
predictions = loaded_model.predict(new_data)

# Interpret the predictions
for i in range(len(new_data_filenames)):
    image_filename = new_data_filenames[i]
    prediction = predictions[i]
    if prediction > 0.5:
        label = 'yes'
    else:
        label = 'no'
    print(f"Image: {image_filename}, Prediction: {label} (Probability: {prediction})")
```
